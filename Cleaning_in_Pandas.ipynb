```python
import pandas as pd 
```

```python
df = pd.read_excel(r'C:\Users\ADMIN\OneDrive\Desktop\DA\Alex_DA_Bootcamp\Python\Resources\Customer Call List.xlsx')
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df = df.drop(columns = 'Not_Useful_Column', errors = 'ignore') 
df
```

```python
df['Last_Name'] = df['Last_Name'].str.strip('123._/') 
df['Last_Name']
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number'] = df['Phone_Number'].replace('[^a-zA-Z0-9]', '', regex = True) 
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) 
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '') 
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '') 
df['Phone_Number'] = df['Phone_Number'].str.replace('--', '') 
df['Phone_Number']
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n = 2, expand = True) 
df
```

```python
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace('Yes', 'Y') 
df[['Paying Customer', 'Do_Not_Contact']] = df[['Paying Customer', 'Do_Not_Contact']].replace('No', 'N') 
df
```

```python
df = df.fillna('') 
df = df.replace('N/a', '')
df
```

```python
for x in df.index: 
    if df.loc[x, 'Phone_Number'] == '': 
        df.drop(x, inplace = True)
df
```

```python
for x in df.index: 
    if df.loc[x, 'Do_Not_Contact'] == 'Y': 
        df.drop(x, inplace = True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('','N') 
df
```
